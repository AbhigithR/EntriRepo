# 5 Find if there is any correlation between age and salary , represent it visually ?

## Calculating the correlation between age and salary

```python
correlation = df['Age'].corr(df['Salary'])

print(f"Correlation between Age and Salary: {correlation}")
```

## Correlation between Age and Salary

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], color='green', alpha=0.7)
plt.title('CORRELATION BETWEEN AGE AND SALARY')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)

# Add a regression line
sns.regplot(x='Age', y='Salary', data=df, scatter=False, color='blue')

plt.show()
```

##### In ABC Company, the age of employees ranges from a minimum of 19 to a maximum of 40. 
##### A positive correlation exists between age and salary,  indicating that, on average, as employees' age increases, their salary tends to rise. 
##### However, it's important to note that we cannot pinpoint a specific rate of increase between age and salary. 
##### The correlation coefficient between age and salary is 0.20, signifying a weak positive
##### correlation, which implies that other factors beyond age play a significant role in determining salary levels.
